```python
import tensorflow as tf

# Define the anchor, positive, and negative samples as tensors
p = tf.constant([0.3388, 0.5373, 0.6385, 0.4345], dtype=tf.float32)
a = tf.constant([0.4000, 0.8000, 0.4000, 0.2000], dtype=tf.float32)
n = tf.constant([0.3383, 0.5220, 0.5340, 0.5727], dtype=tf.float32)

# Define the margin for the triple loss
margin = 0.05

# Compute the triplet loss
positive_distance = tf.reduce_sum(tf.square(a - p))  # Squared L2 distance between anchor and positive
negative_distance = tf.reduce_sum(tf.square(a - n))  # Squared L2 distance between anchor and negative
loss = tf.maximum(0.0, positive_distance - negative_distance + margin)

if (positive_distance + margin) < negative_distance:
  print("easy")

if negative_distance < positive_distance:
  print("herd")

if (positive_distance < negative_distance) and (negative_distance < (positive_distance + margin)):
  print("semi-hard")

# Optionally, you can add a margin to the loss to make it semi-hard triplet loss:
# loss = tf.maximum(0.0, positive_distance - negative_distance + margin)

# Print the computed loss
print("Triple Loss:", loss)
```
